```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
train = pd.read_csv("../data/preprocessing/train.csv")
test = pd.read_csv("../data/preprocessing/test.csv")
```

```python
# test data에는 Credit_Score 변수가 빠져 있으므로 train data에서 데이터를 분리하여 분석
print(len(train.columns), len(test.columns))
```

# 시각화

## 상관분석

```python
train_corr = train.corr(method='pearson')
plt.figure(figsize=(15,6))
sns.heatmap(train_corr, annot=True, cmap='Blues', fmt='.2f')
```

```python
# Annual_Income, Monthly_Inhand_Salary 두 컬럼이 상관관계가 1이므로 둘 중 하나를 제거
# Annual_Income 를 선택
train.drop(columns = 'Monthly_Inhand_Salary', inplace=True)
```

```python
train_corr = train.corr(method='pearson')
plt.figure(figsize=(15,6))
sns.heatmap(train_corr, annot=True, cmap='Blues', fmt='.2f')
```

```python
train_corr.nlargest(5, 'Credit_Score')
```

```python
train_corr.nsmallest(4,'Credit_Score')
```

```python
# 분석할 때 사용할 변수명
# Credit_Mix, Credit_History_Age, Annual_Income, Monthly_Balance, 
# Payment_of_Min_Amount, Delay_from_due_date, Outstanding_Debt, Changed_Credit_Limit
df = train[["Credit_Mix", "Credit_History_Age", "Annual_Income", 
            "Monthly_Balance", "Payment_of_Min_Amount", "Delay_from_due_date", 
            "Outstanding_Debt", "Changed_Credit_Limit",'Credit_Score']]
```

```python
len(df.columns)
```

## histogram and boxplot

```python
for i in df.columns:
    plt.subplot(2,1,1)
    plt.hist(df[i])
    plt.subplot(2,1,2)
    plt.boxplot(df[i], vert=False)
    plt.xlabel(i)
    plt.show()
```
